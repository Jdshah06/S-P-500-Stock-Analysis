```python
import pandas as pd
import scipy as stats
from datetime import datetime

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

s_p_500_data = 'Pandas Data/sandp500/^GSPC.csv'
individual_data = 'Pandas Data/sandp500/all_stocks_5yr.csv'
individual_industry = 'Pandas Data/Industries/data/constituents.csv'

s_p_df = pd.read_csv(s_p_500_data)
individual_df = pd.read_csv(individual_data)
industry_df= pd.read_csv(individual_industry)
```

```python
individual_df.rename(columns = {'Name':'Symbol'}, inplace=True)
stocks_df=individual_df.merge(industry_df, on="Symbol")
stocks_df.head(10)
```

```python
#Maximum Profit booked in a single day?
stocks_df['Daily Change'] = stocks_df['open'] - stocks_df['close']
max_value=stocks_df["Daily Change"].max()
max_day_profit=stocks_df.loc[(stocks_df['Daily Change']== max_value)]
max_day_profit
```

```python
group_by_individual = stocks_df.groupby('Symbol')
daily_max_per_stock=group_by_individual['Daily Change'].max()
stock_daily_max=pd.DataFrame({"Daily Max Per Stock": daily_max_per_stock})
stock_daily_max.head()
```

```python
stock_daily_max = stock_daily_max.reset_index()
stock_daily_max.loc[stock_daily_max['Symbol']=='PCLN']
```

```python
stock_daily_max.sort_values('Daily Max Per Stock', ascending=False)
```

```python
#Maximum Loss Booked in a single Day
min_value= stocks_df["Daily Change"].min()
max_loss_profit=stocks_df.loc[(stocks_df['Daily Change']== min_value)]
max_loss_profit
```

```python
ticker = stocks_df.groupby(['Symbol','date'])[['open','close']].mean()
ticker_df = pd.DataFrame(ticker)
ticker_df
ticker_df = ticker_df.unstack(level=0)
ticker_df = ticker_df.reset_index()
ticker_df
```

```python
# ticker_df['date']= ticker_df['date'].astype('datetime64[ns]')
# ticker_df['date']=pd.to_datetime(ticker_df['date'])
# ticker_df=ticker_df.set_index('date')
# ticker_df.sort_index(inplace=True)

stocks_df['date']= stocks_df['date'].astype('datetime64[ns]')
stocks_df['date']=pd.to_datetime(stocks_df['date'])
weekly_stocks_df=stocks_df.set_index('date')
weekly_stocks_df.sort_index(inplace=True)

def take_first(array_like):
    return array_like[0]

def take_last(array_like):
    return array_like[-1]

output= weekly_stocks_df.resample('W',
                          loffset=pd.offsets.timedelta(days=6))
output= output[['open','high','low','close','volume','Symbol']]
```

```python
output.head()
```

```python
beg_week=output.nth(0)
beg_weekly_change_df=pd.DataFrame(beg_week['open'])
beg_weekly_change_df['Symbol']= beg_week["Symbol"]
beg_weekly_change_df['close']= beg_week['close']
beg_weekly_change_df= beg_weekly_change_df.reset_index()
```

```python
end_week= output.nth(-1)
end_weekly_change_df=pd.DataFrame(end_week['close'])
end_weekly_change_df['Symbol']= end_week['Symbol']
end_weekly_change_df['open']= end_week['open']
end_weekly_change_df = end_weekly_change_df.reset_index()
```
